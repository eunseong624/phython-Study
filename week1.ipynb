```
s1 = set([1,2,3])
print(s1)
l1 = list(s1) # set을 형변환 -> list로 만들어요
l1[0]
```

```
s1 = set([1,2,3])
s2 = set([2,3,4])

print(s1&s2) # &는 교집합을 구해요 -> 반환값은 set이랍니다
print(s1|s2) # |는 합집합을 구해요
print(s1-s2) # -는 차집합을 구해요

s3 = set([1,2,3])
s3.add(4)    # s3에 4를 추가해요
print(s3)

s3.update([5,6,7]) # update는 list의 형태로 만들어요 -> 많은 값을 넣어요
print(s3)

s3.remove(7) # remove는 해당 값을 삭제해요
print(s3)

s3.discard(1) # discard는 해당 값을 삭제해요
s3

s3.clear() # clear는 다 지워버리고 set만 남겨요
s3
```

```
scores = {"국어" : '85', "영어" : '90', "수학" : '95'}

for subject, score in scores.items(): # 변수 두개도 올 수 있어요
  # print(subject , score)
  print(f"{subject} {score}") # f를 이용해야해요
```
